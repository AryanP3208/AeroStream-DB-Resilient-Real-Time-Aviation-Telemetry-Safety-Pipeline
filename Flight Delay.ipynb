```python
import pandas as pd
import os

file_path = os.path.join(path, "flight_data_2024.csv")  # Corrected filename
df = pd.read_csv(file_path)

print("Shape:", df.shape)
df.head()
```

```python
df["cancelled"].value_counts()
```

```python
cancel_rate = df["cancelled"].mean() * 100
print("Cancellation Rate:", cancel_rate, "%")
```

```python
import matplotlib.pyplot as plt

counts = df["cancelled"].value_counts()

plt.figure()
plt.pie(counts, labels=["Not Cancelled", "Cancelled"], autopct="%1.1f%%")
plt.title("Flight Cancellation Distribution")
plt.show()
```

```python
df["weather_delay"].describe()
```

```python
df["weather_impact"] = df["weather_delay"].apply(lambda x: 1 if x > 0 else 0)

df["weather_impact"].value_counts(normalize=True) * 100
```

```python
import matplotlib.pyplot as plt

counts = df["weather_impact"].value_counts()

plt.figure()
plt.pie(counts, labels=["No Weather Delay", "Weather Delay"], autopct="%1.1f%%")
plt.title("Weather Delay Impact Distribution")
plt.show()
```

```python
def weather_category(x):
    if x == 0:
        return "No Delay"
    elif x <= 30:
        return "Minor"
    elif x <= 120:
        return "Moderate"
    else:
        return "Severe"

df["weather_category"] = df["weather_delay"].apply(weather_category)

df["weather_category"].value_counts()
```
